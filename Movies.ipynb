## Data Manipulation


```python
# select the 2nd to 4 character to get the first year of the year column
df['New_Year'] = df['YEAR'].str[1:5]
df['New_Year'].head()
```

```python
df.head(10)
```

```python
# new year still has unwanted values
df['New_Year'].unique()
```

```python
# filter for values that only starts with 1 or 2
df = df[df['New_Year'].str.startswith(('1', '2'), na = False)]
```

```python
df['New_Year'].unique()
```

### Filling Missing Data with Grouped Statistics

```python
# Find sum of null values
df.isna().sum()
```

```python
# fill in the genre with with mode
df['GENRE'] = df['GENRE'].fillna(df['GENRE'].mode()[0])
```

```python
# find unique values
df['GENRE'].unique()
```

```python
# Create genre category
genre_cat = ['Animation', 'Horror', 'Thriller', 'Comedy', 'Action', 'Drama']
```

```python
# group the genre into these categories 
drama = "Drama|Romance|Mystery|Biography"
action = "Action|Sci-Fi|Crime"
anime = "Animation"
horror = "Horror"
thriller = "Thriller|History"
comedy = "Comedy"
```

```python
# create a condtion list using .contains
conditions = [(df['GENRE'].str.contains(anime)), 
              (df['GENRE'].str.contains(horror)), 
              (df['GENRE'].str.contains(thriller)), 
              (df['GENRE'].str.contains(comedy)), 
              (df['GENRE'].str.contains(action)),
              (df['GENRE'].str.contains(drama))]
```

### np.select

#### creates a column based on values containing in another columns

```python
df['genre'] = np.select(conditions, genre_cat, default = "other")
```

```python
df['genre'].unique()
```

```python
# remove "," from the votes column
df['VOTES'] = df['VOTES'].str.replace(',', '')
```

```python
# change datatype to float
df['VOTES'] = df['VOTES'].astype(float)
```

```python
print(df['VOTES'].describe())
```

### .median().to_dict()

#### Grouping and agg the stats to dict, fill missing values with the goruped stats

```python

vot_dic = df.groupby('genre')['VOTES'].median().to_dict()
df['VOTES'] = df['VOTES'].fillna(df['genre'].map(vot_dic))
```

### pd.cut

#### creating a category based on numeric values... 

```python
# find the inter - quantile ranges 
twenty_f = df['VOTES'].quantile(0.25)
m = df['VOTES'].median()
seventy_f = df['VOTES'].quantile(0.75)
maxi = df['VOTES'].max()
```

```python
# creates cats
labels = ['Low', 'Normal', 'High', 'Very High']
```

```python
# created bins based on the labels
bins = [0, twenty_f, m, seventy_f, maxi]
```

```python
df['votes_cat'] = pd.cut(df['VOTES'], labels = labels, bins = bins)
```

```python
df["votes_cat"].head()
```

```python
sns.countplot(data = df, x = "votes_cat")
plt.show()
```

```python
df.info()
```

```python
# replasing unwanted characters in Gross and converting type to float
df['Gross'] = df['Gross'].str.replace("$", "").str.replace("M", "").astype("float")
```

```python
df['Gross'].unique()
```

```python
df = df.round(2)
```

```python
df.describe()
```

### Drop unwanted columns 

```python
df.columns
```

```python
df = df.drop(['YEAR', 'GENRE', 'ONE-LINE', 'STARS'], axis = 1)
```

```python
df.reset_index()
```

```python
# find remain null columns
df.isna().sum()
```

```python
# fill na in rating with grouped sum stats
rat_dic = df.groupby('genre')['RATING'].median().to_dict()
df['RATING'] = df['RATING'].fillna(df['genre'].map(rat_dic))

```

```python
# create rating cats
twenty_f = df['RATING'].quantile(0.08)
m = df['RATING'].quantile(0.50)
seventy_f = df['RATING'].quantile(0.75)
maxi = df['RATING'].max()

labels = ['Low', 'Normal', 'High', 'Very High']

bins = [0, twenty_f, m, seventy_f, maxi]

df['rating_cat'] = pd.cut(df['RATING'], labels = labels, bins = bins)
```

```python
# fill runtime with group sum stats
time_dic = df.groupby('genre')['RunTime'].median().to_dict()
df['RunTime'] = df['RunTime'].fillna(df['genre'].map(time_dic))
```

### Outliers

#### IQR = 75th - 25th percentile
#### Upper Outlier > 75th + (1.5 * IQR)

```python
# Identifying thresholds

seventy_fifth = df['RunTime'].quantile(0.75)
twenty_fifth = df['RunTime'].quantile(0.25)

print(seventy_fifth, twenty_fifth)
```

```python
IQR = seventy_fifth - twenty_fifth
IQR
```

```python
#upper threshold
upper = seventy_fifth + (1.5 * IQR)

#lower threshold
lower = twenty_fifth - (1.5 * IQR)

print(upper, lower)
```

```python
outliers = (df['RunTime'] < lower) | (df['RunTime'] > upper)
run_time_outliers = df[outliers]
df3 = df[~outliers]
df3.shape
```

```python
sns.boxplot(data = df3, x = 'genre', y = 'RunTime')
plt.show()
```
